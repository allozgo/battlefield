```python
def barco(tablero01, tablero31, size):
    def separa_barco(tablero41, cambio, orientacion1):
        x = cambio[0]
        y = cambio[1]
        tablero41[x,y] = "O"
        if orientacion1 != "S":
            if (x+1) <= 9:
                tablero41[x+1, y] = "O"
        if orientacion1 != "N":
            if (x-1) >= 0:
                tablero41[x-1, y] = "O"
        if orientacion1 != "E":
            if (y+1) <= 9:
                tablero41[x,y+1] = "O"
        if orientacion1 != "O":
            if (y-1) >= 0:
                tablero41[x,y-1] = "O"
        if (x-1) >= 0 and (y-1) >= 0:
            tablero41[x-1, y-1] = "O"
        if (x-1) >= 0 and (y+1) <= 9:
            tablero41[x-1, y+1] = "O"
        if (x+1) <= 9 and (y-1) >= 0:
            tablero41[x+1, y-1] = "O"
        if (x+1) <= 9 and (y+1) <= 9:
            tablero41[x+1, y+1] = "O"
        return tablero41

    def poner_barco(tablero02, tablero32, posicion, orientacion, tamano):
        pos_barco = [list(posicion)]
        tablero2 = tablero02.copy()
        tablero4 = tablero32.copy()
        cambio = list(posicion)
        tablero2[posicion]="O"
        tablero4 = separa_barco(tablero4, cambio, orientacion)
        if tamano == 1:
            cambio = list(posicion)
            pos_barco = [list(posicion)]
            tablero2[posicion]="O"
            tablero4 = separa_barco(tablero4, cambio, 1)
            return tablero2, tablero4, pos_barco
        if orientacion == "N":
            for i in range(tamano-1):
                cambio[0] -= 1
                if cambio[0] < 0:
                    pos_barco = []
                    posicion = tuple(np.random.randint(0, 10, (2)))
                    orientacion = np.random.choice(["N", "S", "E", "O"])
                    tablero2, tablero4, pos_barco = poner_barco(tablero02, tablero32, posicion, orientacion, tamano)
                    return tablero2, tablero4, pos_barco
                if tablero4[tuple(cambio)] == "O":
                    pos_barco = []
                    posicion = tuple(np.random.randint(0, 10, (2)))
                    orientacion = np.random.choice(["N", "S", "E", "O"])
                    tablero2, tablero4, pos_barco = poner_barco(tablero02, tablero32, posicion, orientacion, tamano)
                    return tablero2, tablero4, pos_barco
                tablero2[tuple(cambio)] = "O"
                pos_barco.append(cambio.copy())
                if i == (tamano - 2):
                    tablero4 = separa_barco(tablero4, cambio, 1)
                else:
                    tablero4 = separa_barco(tablero4, cambio, orientacion)
            return tablero2, tablero4, pos_barco
        if orientacion == "S":
            for i in range(tamano-1):
                cambio[0] += 1
                if cambio[0] > 9:
                    pos_barco = []
                    posicion = tuple(np.random.randint(0, 10, (2)))
                    orientacion = np.random.choice(["N", "S", "E", "O"])
                    tablero2, tablero4, pos_barco = poner_barco(tablero02, tablero32, posicion, orientacion, tamano)
                    return tablero2, tablero4, pos_barco
                if tablero4[tuple(cambio)] == "O":
                    pos_barco = []
                    posicion = tuple(np.random.randint(0, 10, (2)))
                    orientacion = np.random.choice(["N", "S", "E", "O"])
                    tablero2, tablero4, pos_barco = poner_barco(tablero02, tablero32, posicion, orientacion, tamano)
                    return tablero2, tablero4, pos_barco
                tablero2[tuple(cambio)] = "O"
                pos_barco.append(cambio.copy())
                if i == (tamano - 2):
                    tablero4 = separa_barco(tablero4, cambio, 1)
                else:
                    tablero4 = separa_barco(tablero4, cambio, orientacion)
            return tablero2, tablero4, pos_barco
        if orientacion == "E":
            for i in range(tamano-1):
                cambio[1] += 1
                if cambio[1] > 9:
                    pos_barco = []
                    posicion = tuple(np.random.randint(0, 10, (2)))
                    orientacion = np.random.choice(["N", "S", "E", "O"])
                    tablero2, tablero4, pos_barco = poner_barco(tablero02, tablero32, posicion, orientacion, tamano)
                    return tablero2, tablero4, pos_barco
                if tablero4[tuple(cambio)] == "O":
                    pos_barco = []
                    posicion = tuple(np.random.randint(0, 10, (2)))
                    orientacion = np.random.choice(["N", "S", "E", "O"])
                    tablero2, tablero4, pos_barco = poner_barco(tablero02, tablero32, posicion, orientacion, tamano)
                    return tablero2, tablero4, pos_barco
                tablero2[tuple(cambio)] = "O"
                pos_barco.append(cambio.copy())
                if i == (tamano - 2):
                    tablero4 = separa_barco(tablero4, cambio, 1)
                else:
                    tablero4 = separa_barco(tablero4, cambio, orientacion)
            return tablero2, tablero4, pos_barco
        if orientacion == "O":
            for i in range(tamano-1):
                cambio[1] -= 1
                if cambio[1] < 0:
                    pos_barco = []
                    posicion = tuple(np.random.randint(0, 10, (2)))
                    orientacion = np.random.choice(["N", "S", "E", "O"])
                    tablero2, tablero4, pos_barco = poner_barco(tablero02, tablero32, posicion, orientacion, tamano)
                    return tablero2, tablero4, pos_barco
                if tablero4[tuple(cambio)] == "O":
                    pos_barco = []
                    posicion = tuple(np.random.randint(0, 10, (2)))
                    orientacion = np.random.choice(["N", "S", "E", "O"])
                    tablero2, tablero4, pos_barco = poner_barco(tablero02, tablero32, posicion, orientacion, tamano)
                    return tablero2, tablero4, pos_barco
                tablero2[tuple(cambio)] = "O"
                pos_barco.append(cambio.copy())
                if i == (tamano - 2):
                    tablero4 = separa_barco(tablero4, cambio, 1)
                else:
                    tablero4 = separa_barco(tablero4, cambio, orientacion)
            return tablero2, tablero4, pos_barco
    posicion = tuple(np.random.randint(0, 10, (2)))
    while tablero31[posicion] == "O":
        posicion = tuple(np.random.randint(0, 10, (2)))
    tablero01[posicion] == "O"
    orientacion = np.random.choice(["N", "S", "E", "O"])
    tablero01, tablero31, pos_barco = poner_barco(tablero01, tablero31, posicion, orientacion, size)
    return tablero01, tablero31, pos_barco
```

```python
def tablero_barco():
    tablero = np.full((10,10), " ")
    tablero3 = np.full((10,10), " ")
    tablero0 = tablero.copy()
    tablero6 = tablero3.copy()
    barcos = {}
    for i in range(4):
        eslora = 1
        tablero0, tablero6, p_barco = barco(tablero0, tablero6, eslora)
        barcos["barco"+str(i)] = p_barco
    for i in range(4,7):
        eslora = 2
        tablero0, tablero6, p_barco = barco(tablero0, tablero6, eslora)
        barcos["barco"+str(i)] = p_barco
    for i in range(7,10):
        eslora = 3
        tablero0, tablero6, p_barco = barco(tablero0, tablero6, eslora)
        barcos["barco"+str(i)] = p_barco
    tablero0, tablero6, p_barco = barco(tablero0, tablero6, 4)
    barcos["barco10"] = p_barco
    return tablero0, barcos
```

```python
tablero_tiro_jugador = np.full((10,10), " ")
tablero_tiro_maquina = np.full((10,10), " ")

def tirada_jugador(tablero_jugador1, tablero_tiro_jugador1, tablero_maquina1, barcos_maquina1):
    tiro = True
    while tiro:
        tiro_jugador_x = int(input("Dame la coordenada x"))
        if (tiro_jugador_x < 0) or (tiro_jugador_x > 9):
            print("Tirada errónea, tira de nuevo")
            tiro_jugador_x = int(input("Dame la coordenada x"))
        tiro_jugador_y = int(input("Dame la coordenada y"))
        if (tiro_jugador_y < 0) or (tiro_jugador_y > 9):
            print("Tirada errónea, tira de nuevo")
            tiro_jugador_y = int(input("Dame la coordenada y"))
        tiro_jugador = (tiro_jugador_x, tiro_jugador_y)
        if tablero_maquina1[tiro_jugador] == 'O':
            tablero_maquina1[tiro_jugador] = 'X'
            tablero_tiro_jugador1[tiro_jugador] = 'X'
            print("Has acertado un barco, continua")
            for barco in barcos_maquina1:
                for i in range(len(barco)):
                    if barco[i] == list(tiro_jugador):
                        barco.pop(i)

        else:
            tablero_tiro_jugador1[tiro_jugador] = 'A'
            print("Has hecho agua")
            tiro = False
    print("Tablero de la máquina")
    print(tablero_maquina1)
    print("Tablero de tiros del jugador")
    print(tablero_tiro_jugador1)
    return tablero_jugador1, tablero_tiro_jugador1, tablero_maquina1, barcos_maquina1
```
